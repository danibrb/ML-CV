```python
import numpy as np
import matplotlib.pyplot as plt

# Definizione della funzione di vincolo
def lorentzian_constraint(s_sq, alpha=1.0):
    return alpha * (1.0 / (1.0 + (s_sq - 1.0)**2))

# Generazione dei valori per s barrato al quadrato (s^2)
# Campioniamo da 0 a 2 per vedere la simmetria attorno a 1
s_sq_values = np.linspace(-0.5, 2.5, 500)
f_values = lorentzian_constraint(s_sq_values, alpha=1.0)

# Creazione del grafico
plt.figure(figsize=(8, 5))
plt.plot(s_sq_values, f_values, label=r'$f(\bar{s}^2) = \alpha \frac{1}{1 + (\bar{s}^2 - 1)^2}$', color='royalblue', linewidth=2.5)

# Evidenziazione del punto di massimo target (s^2 = 1)
plt.axvline(x=1.0, color='red', linestyle='--', alpha=0.7, label=r'Target $\bar{s}^2 = 1$')
plt.scatter([1.0], [1.0], color='red', zorder=5)

# Configurazione degli assi e dettagli estetici
plt.title(r'Funzione di Vincolo Lorentziana (per $\alpha = 1$)', fontsize=14, pad=15)
plt.xlabel(r'$\bar{s}^2$', fontsize=13)
plt.ylabel(r'$f(\bar{s}^2)$', fontsize=13)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=11, loc='upper right')

# Mostra il plot
plt.tight_layout()
plt.show()
```
